# Main function

```python
# now we know there are 20 rows in 1 page and there are 500 pages ( as seen from online json viewer)
# so we'll create a loop to fetch all pages and concatenate the results in one DataFrame, later convert it to a CSV file

# starting with an empty DataFrame which will be used to store the results
df = pandas.DataFrame()
```

```python
# empty list to store DataFrames for each page
df_list = []
# Looping through the pages to fetch all top-rated movies 1 - 501 
for i in range (1,501):
    response =  requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=<<API KEY>>&page={}".format(i))
    # temp_df creats a DataFrame for each page and selects specific columns of results and appends it to df_list
    temp_df = pandas.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df_list.append(temp_df)
# Concatenating all DataFrames collected in df_list into a single DataFrame
df= pandas.concat(df_list, ignore_index=True)
    
```

```python
# Saving the DataFrame to a CSV file
df.to_csv('top_rated_movies.csv', index=False)
```
